```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import t
```

```python
# -----------------------------
# Given Data # with the 1 MHz, 1000x TIA PD
# -----------------------------
Vppin_dBm = 7  # input power in dBm
RF_Signal_MHz = np.array([1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5])
Vppout_mV = np.array([105, 80, 67, 55, 46, 40, 33, 31])
```

```python
# -----------------------------
# Convert dBm -> Vpp (mV)
# Assumptions:
# - 50 ohm system
# - sinusoidal signal
# -----------------------------
R = 50  # ohms

# Convert dBm to Watts
P_mW = 10**(Vppin_dBm / 10)
P_W = P_mW / 1000

# Convert Power -> Vrms
Vrms = np.sqrt(P_W * R)

# Convert Vrms -> Vpp for sinusoid
Vppin_V = 2 * np.sqrt(2) * Vrms

# Convert to mV
Vppin_mV = Vppin_V * 1e3

print(f"Input Voltage: {Vppin_mV:.2f} mVpp")

# -----------------------------
# Calculate Loss (Voltage dB)
# Gain = 20 log10(Vin/Vout)
# -----------------------------
Gain_dB = 20 * np.log10(Vppout_mV / Vppin_mV)

plt.figure()
plt.plot(RF_Signal_MHz, Gain_dB, marker='o')
plt.xlabel("RF Signal (MHz)")
plt.ylabel("Gain (dB)")
plt.title("Voltage Gain vs RF Frequency")
plt.grid(True)
plt.show()
```

```python
"""
Fit a single-pole magnitude model to Gain(dB) vs frequency data,
extrapolate to lower frequencies, and show 95% CI from covariance.

Model (voltage ratio):
    G_lin(f) = G0 / sqrt(1 + (f/fc)**2)
Gain in dB:
    G_db(f) = 20*log10(G_lin(f))

Inputs: arrays RF_MHz (MHz), Gain_dB (dB)

Outputs: best-fit G0 (linear), fc (MHz), plot saved to disk, printed uncertainties.
"""



RF_MHz = RF_Signal_MHz


# --- model definitions ---
def G_lin_model(f_mhz, G0_lin, fc_mhz):
    """Voltage linear gain model: G0 / sqrt(1 + (f/fc)^2)"""
    f = f_mhz
    return G0_lin / np.sqrt(1.0 + (f / fc_mhz)**2)

def G_db_model(f_mhz, G0_lin, fc_mhz):
    """Return model in dB"""
    return 20.0 * np.log10(G_lin_model(f_mhz, G0_lin, fc_mhz))

# --- initial guesses ---
# G0 guess: convert highest (lowest loss) dB back to linear as approximate plateau
approx_G0_lin = 10**(np.max(Gain_dB) / 20.0)  # peak measured (closest to low-f plateau)
approx_fc = 2.0  # MHz, guess in-band corner within measured range (tweakable)

p0 = [approx_G0_lin, approx_fc]

# Fit in dB space (weights optional). Use absolute dB residuals since instrument noise often approx Gaussian in voltage, but fitting in dB is common.
popt, pcov = curve_fit(G_db_model, RF_MHz, Gain_dB, p0=p0, maxfev=10000)
G0_hat, fc_hat = popt
perr = np.sqrt(np.diag(pcov)) # standard errors for parameters (in same units as parameters)

# convert parameter errors to approximate dB uncertainty for G0:
G0_err = perr[0]
fc_err = perr[1]

# Compute fitted curve and CI on model predictions (approx using linearized error propagation)
f_plot = np.logspace(np.log10(RF_MHz.min()/10), np.log10(RF_MHz.max()*10), 400)  # extend beyond measured range
G_fit_db = G_db_model(f_plot, G0_hat, fc_hat)

# Compute Jacobian matrix at each f for propagation: J = dG_db/dtheta (theta = [G0_lin, fc])
# Analytical derivatives:
def dGdb_dG0(f_mhz, G0_lin, fc_mhz):
    # d/dG0 of 20*log10( G0 / sqrt(1 + (f/fc)^2) ) = 20 / (ln(10)) * (1 / (G0 * sqrt(...)))
    denom = np.sqrt(1.0 + (f_mhz / fc_mhz)**2)
    return 20.0 / np.log(10) * (1.0 / (G0_lin * denom))

def dGdb_dfc(f_mhz, G0_lin, fc_mhz):
    # derivative w.r.t fc (more complex)
    denom = np.sqrt(1.0 + (f_mhz / fc_mhz)**2)
    # d/d(fc) of log10(G0/denom) = -1/(ln(10)) * (1/denom) * d(denom)/d(fc) / (G0/denom) * 20
    # Better to compute derivative of G_lin then chain-rule; we derive numerically for stability:
    eps = 1e-6 * max(1.0, fc_mhz)
    g_plus = 20.0 * np.log10(G_lin_model(f_mhz, G0_lin, fc_mhz + eps))
    g_minus = 20.0 * np.log10(G_lin_model(f_mhz, G0_lin, fc_mhz - eps))
    return (g_plus - g_minus) / (2 * eps)

# Build covariance of predicted G_db via linear propagation: var_y = J * cov_theta * J^T
J = np.vstack((
    dGdb_dG0(f_plot, G0_hat, fc_hat),
    dGdb_dfc(f_plot, G0_hat, fc_hat)
)).T  # shape (npoints, 2)
# predicted variance at each freq:
pred_var = np.einsum('ij,jk,ik->i', J, pcov, J)  # J * cov * J^T diagonal
pred_std = np.sqrt(np.maximum(pred_var, 0.0))

# 95% CI (approx, using t quantile for small N)
dof = max(1, len(RF_MHz) - len(p0))
tval = t.ppf(0.975, dof)
upper = G_fit_db + tval * pred_std
lower = G_fit_db - tval * pred_std

# --- print results ---
print(f"Best-fit low-frequency linear gain G0 = {G0_hat:.4e} (V/V)  ± {G0_err:.3e} (1 sigma)")
print(f"Best-fit corner frequency fc = {fc_hat:.4f} MHz  ± {fc_err:.4f} MHz (1 sigma)")
print(f"Equivalent low-frequency gain in dB: {20*np.log10(G0_hat):.3f} dB")

# --- plot measured, fit, extrapolation with 95% CI ---
plt.figure(figsize=(7,4))
plt.semilogx(RF_MHz, Gain_dB, 'o', label='Measured')
plt.semilogx(f_plot, G_fit_db, '-', label='Single-pole fit (extrapolated)')
plt.fill_between(f_plot, lower, upper, color='gray', alpha=0.3, label='Approx. 95% CI')
plt.xlabel('Frequency (MHz)')
plt.ylabel('Gain (dB)')
plt.title('Single-pole fit and extrapolation to lower frequency')
plt.legend(fontsize='small')
plt.grid(True, which='both', ls=':')

plt.xlim([f_plot.min(), f_plot.max()])
plt.ylim([min(lower.min(), Gain_dB.min())-3, max(upper.max(), Gain_dB.max())+3])
plt.tight_layout()
plt.savefig('single_pole_extrapolation.png', dpi=200)
print("Saved plot to single_pole_extrapolation.png")
```
